# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

def ReLU(x):
    return (x * (x > 0))

def dReLU(x):
    return (1 * (x > 0))

x = plt.linspace(-4, 4, 100)

plt.grid()



plt.title('ReLU Function')

#plt.text(-5, 8, r'$ReLU(x)=max{(0,x)}$', fontsize=15)

plt.plot(x, ReLU(x), 'b')

plt.plot(x, dReLU(x), 'g')

plt.legend(['ReLU', 'dReLU'], loc='upper left')

plt.show()
```
